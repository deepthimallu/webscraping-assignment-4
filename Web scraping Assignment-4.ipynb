```python
#getting the url
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
time.sleep(2)
#Getting title of the show
Name=[]
try:
    title=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a')
    for i in title:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')
#Getting year span
year=[]
try:
    span=driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
    for i in span:
        year.append(i.text)
except NoSuchElementException:
    year.append('-')
#Getting Genre
genre=[]
try:
    Genre=driver.find_elements_by_xpath('//span[@class="genre"]')
    for i in Genre:
        genre.append(i.text)
except NoSuchElementException:
    genre.append('-')
#Getting Ratings
rating=[]
try:
    Rate=driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]/span')
    for i in Rate:
        rating.append(i.text)
except NoSuchElementException:
    rating.append('-')
Rating=[]
for i in range(0,len(rating)-1,2):
    Rating.append(rating[i+1])
#Getting Runtime details
runtime=[]
try:
    Run=driver.find_elements_by_name('nv')
    for i in Run:
        runtime.append(i.text)
except NoSuchElementException:
    runtime.append('-')
#getting votes
votes=[]
try:
    Votes=driver.find_elements_by_xpath('//span[@class="runtime"]')
    for i in Votes:
        votes.append(i.text)
except NoSuchElementException:
    votes.append('-')

# Getting all in a DataFrame
tv_series=pd.DataFrame({})
tv_series['Name']=Name[:100]
tv_series['Year Span']=year[:100]
tv_series['Genre']=genre[:100]
tv_series['Run Time']=runtime[:100]
tv_series['Ratings']=Rating[:100]
tv_series['Votes']=votes[:100]
tv_series   
```
